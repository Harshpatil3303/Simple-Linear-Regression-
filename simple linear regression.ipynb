```python
import pandas as pd

data = pd.read_csv("/content/delivery_time (1).csv")
data.head()
```

```python
data.corr()
```

```python
import seaborn as sns
sns.displot(data['Delivery Time'],kde=True)
```

```python
sns.displot(data['Sorting Time'],kde=True)
```

```python
import statsmodels.formula.api as smf
```

```python
df1 = data.rename({'Delivery Time':'Delivery_Time','Sorting Time':'Sorting_Time'}, axis = 1)
df1
```

```python
model = smf.ols('Delivery_Time~Sorting_Time', data = df1).fit()
```

```python
model.summary()
```

```python
sns.regplot(x="Sorting_Time", y="Delivery_Time", data=df1);
```

```python
model.params
```

```python
print(model.tvalues,'\n', model.pvalues)
```

```python
(model.rsquared,model.rsquared_adj)
```
